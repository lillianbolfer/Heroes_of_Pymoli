## Purchasing Analysis (Age)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
purchases_by_SN = purchase_data_df.groupby('SN')

total_purchases=purchases_by_SN["SN"].count()

total_purchase_value= pd.to_numeric(purchases_by_SN["Price"].sum())

average_price= total_purchase_value/total_purchases

summary_table_spenders_df= pd.DataFrame({"Purchase Count": total_purchases, "Average Purchase Price": average_price, 
                                     "Total Purchase Value": total_purchase_value})
summary_table_spenders_df = summary_table_spenders_df.sort_values("Total Purchase Value", ascending=False)

summary_table_spenders_df.head()
```
